<h1>Loading Data

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("breast_cancer.xls")
data = data.values
x = data[:, 2:-2]
y = data[:, 1:2]
```

<h1>Encoding M/B as 1/0 where 1 means Malignant and 0 means Benign

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("", OneHotEncoder(), [0])], remainder="passthrough")
y = ct.fit_transform(y)
y = y[:, 1]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=29)
```

<h1>Prediction using SVM model
<h2>and plotting of confusion matrix for Train/Test data using models based on various kernels

```python
from sklearn.svm import SVC
```

<h1>Using Linear kernel

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
svn_classifier = SVC(kernel="linear")
svn_classifier.fit(x_train, y_train)
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(
    svn_classifier, x_train, y_train, ax=ax, cmap=plt.cm.Reds)
ax.set(xlabel='Predicted', ylabel='Actual',
       title='Confusion Matrix (Train Data)')
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(
    svn_classifier, x_test, y_test, ax=ax, cmap=plt.cm.Blues)
ax.set(xlabel='Predicted', ylabel='Actual',
       title='Confusion Matrix (Test Data)')
```

<h1>Using Sigmoid kernel

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
svn_classifier = SVC(kernel="sigmoid")
svn_classifier.fit(x_train, y_train)
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(
    svn_classifier, x_train, y_train, ax=ax, cmap=plt.cm.Purples)
ax.set(xlabel='Predicted', ylabel='Actual',
       title='Confusion Matrix (Train Data)')
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(
    svn_classifier, x_test, y_test, ax=ax, cmap=plt.cm.Oranges)
ax.set(xlabel='Predicted', ylabel='Actual',
       title='Confusion Matrix (Test Data)')
```

<h1>Using RBF kernel

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
svn_classifier = SVC(kernel="rbf")
svn_classifier.fit(x_train, y_train)
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(
    svn_classifier, x_train, y_train, ax=ax, cmap=plt.cm.Greens)
ax.set(xlabel='Predicted', ylabel='Actual',
       title='Confusion Matrix (Train Data)')
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(
    svn_classifier, x_test, y_test, ax=ax, cmap=plt.cm.YlOrBr)
ax.set(xlabel='Predicted', ylabel='Actual',
       title='Confusion Matrix (Test Data)')
```

<h1>Using POLY kernel

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
svn_classifier = SVC(kernel="poly", degree=3)
svn_classifier.fit(x_train, y_train)
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(
    svn_classifier, x_train, y_train, ax=ax, cmap=plt.cm.GnBu)
ax.set(xlabel='Predicted', ylabel='Actual',
       title='Confusion Matrix (Train Data)')
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(
    svn_classifier, x_test, y_test, ax=ax, cmap=plt.cm.PuRd)
ax.set(xlabel='Predicted', ylabel='Actual',
       title='Confusion Matrix (Test Data)')
```
